```python
# -*- coding: utf-8 -*-

import folium

# 地図生成
folium_map = folium.Map(location=[35.790762308802826, 139.92295550958076],zoom_start=15)

# マーカープロット（ポップアップ設定，色変更，アイコン変更）
folium.Marker(
    location=[35.798845199201956, 139.92677588074534],
    popup='matsudoshinden Station',
    icon=folium.Icon(color='purple', icon='home')
).add_to(folium_map)

# 地図表示
folium_map
```

```python
# -*- coding: utf-8 -*-

import folium

# 地図生成
folium_map = folium.Map(location=[35.798845199201956, 139.92677588074534],zoom_start=15)

# サークルマーカープロット(外径色は赤，内径色は青)
folium.CircleMarker(
    location=[35.798845199201956, 139.92677588074534],
    radius=40,
    color='#ff0000',
    fill_color='#0000ff'
).add_to(folium_map)

# 地図表示
folium_map
```
