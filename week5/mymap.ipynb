```python
import folium
import pandas as pd

# データ格納（名前，緯度経度）
df = pd.DataFrame({
    'P': ['(有)光ヶ丘駐車場', 'タイムズB', 'ナビパーク光ヶ丘団地第一駐車場', '(有)光ヶ丘駐車場', 'グリーンタウン光が丘 来客者用駐車場'],
    'latitude': [35.83256084163083,35.837859027062365,35.83140455502312,35.834420130810294,35.82980559620053,],
    'longtude': [139.95678221107946,139.95601556112456, 139.9569773536314,139.95968732303604,139.95745439170176],
})

# 地図生成（麗澤大学中心）
folium_map = folium.Map(location=[35.83342361090082, 139.95488776231124], zoom_start=15)
folium.Marker(
    location=[35.83342361090082, 139.95488776231124],
    popup="Reitaku University, Private University!",
    tooltip="Reitaku University",
    ).add_to(folium_map)

folium.CircleMarker(
    location=[35.83342361090082, 139.95488776231124],
    radius=100,
    color='#ff0000',
    fill_color='#0000ff'
).add_to(folium_map)

# マーカープロット
for i, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longtude']],
        popup=row['P'],
        icon=folium.Icon(color='red')
    ).add_to(folium_map)

# 地図表示
folium_map
```

```python
folium_map.save("map.html")
```
