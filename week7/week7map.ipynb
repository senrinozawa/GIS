```python
import geopandas as gpd
import pandas as pd
```

```python
# データを変数に入れる
ibaraki_ku = gpd.read_file('ibaraki_ku.geojson')
```

```python
# とりあえずマップとしてアウトプット
ibaraki_ku.plot()
```

```python
# 最初の５列は？
ibaraki_ku.head()
```

```python
# データの情報
ibaraki_ku.info()
```

```python
ibaraki_pop = pd.read_csv('ibaraki.csv')
```

```python
ibaraki_pop.head()
```

```python
ibaraki_pop.info()
```

```python
# rename
ibaraki_ku.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
```

```python
# check the new fieldname
ibaraki_ku
```

```python
# OBJECTからintに変える
ibaraki_ku['KEY_CODE'] = ibaraki_ku['KEY_CODE'].astype(int)
```

```python
# checkする
ibaraki_ku.info()
```

```python
# ここでようやく merge コマンドで統合させて、新しい変数に入れる
ibaraki_ku_pop = ibaraki_ku.merge(ibaraki_pop,on='KEY_CODE')
```

```python
# check!
ibaraki_ku_pop
```

```python
ibaraki_ku_pop.info()
```

```python
# change from OBJECT to int
ibaraki_ku_pop["T001081001"] = ibaraki_ku_pop["T001081001"].astype(int)
```

```python
# check
ibaraki_ku_pop.info()
```

```python
ibaraki_ku_pop.plot(column='T001081001',legend=True)
```

```python
ibaraki_ku_pop.plot(figsize=(10,10),
                  column='T001081001',
                  legend=True)
```

```python
ibaraki_ku_pop.plot(figsize=(10,10),
                  column='T001081001',
                  legend=True,
                  cmap='RdYlGn_r')
```
